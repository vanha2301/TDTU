```python
#ex1.1
import cv2
import numpy as np
from matplotlib import pyplot as plt

img1 = cv2.imread("input1.jpg")
img2 = cv2.imread("input2.jpg")

# tự động tùy chỉnh kích thức 2 ảnh về cùng 1 kích thước
if img1.shape != img2.shape:
    img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

#edit line dọc center two image
line_thickness = 5
height = img1.shape[0]
width = img1.shape[1]
line = np.full((height, line_thickness, 3), (0, 0, 255), dtype=np.uint8)


side_by_side = np.hstack((img1, line, img2))

size_by_size_rgb = cv2.cvtColor(side_by_side, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12, 6))
plt.imshow(side_by_side_rgb)
plt.title('input1 input2')
plt.axis('off')
plt.show()
```

```python
added = np.add(img1, img2)
added_rgb = cv2.cvtColor(added, cv2.COLOR_BGR2RGB)
# plt.figure(figsize=(6, 6))
plt.imshow(added_rgb)
plt.title('added')
plt.show()
```

```python
weighted = cv2.addWeighted(img1, 0.7, img2, 0.3, 0)
weighted_rgb = cv2.cvtColor(weighted, cv2.COLOR_BGR2RGB)

plt.imshow(weighted_rgb)
plt.title('Weighted Addition (cv2.addWeighted)')
plt.axis('off')
plt.show()
```

```python
import cv2

cap = cv2.VideoCapture(0)

if (cap.isOpened() == False):
    print("Error opening video stream or file")
# importing libraries
import cv2

# Create a VideoCapture object and read from input file
cap = cv2.VideoCapture('Spend Your Summer Vacations\
Wisely! Ft. Sandeep Sir _ GeeksforGeeks.mp4')

# Check if camera opened successfully
if (cap.isOpened()== False):
    print("Error opening video file")

# Read until video is completed
while(cap.isOpened()):
    
# Capture frame-by-frame
    ret, frame = cap.read()
    if ret == True:
    # Display the resulting frame
        cv2.imshow('Frame', frame)
        
    # Press Q on keyboard to exit
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break

# Break the loop
    else:
        break

# When everything done, release
# the video capture object
cap.release()

# Closes all the frames
cv2.destroyAllWindows()
```
